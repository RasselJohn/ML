# Подход 1: градиентный бустинг "в лоб"
5. Забудем, что в выборке есть категориальные признаки,
и попробуем обучить градиентный бустинг над деревьями на имеющейся матрице "объекты-признаки".
Зафиксируйте генератор разбиений для кросс-валидации по 5 блокам (KFold),
не забудьте перемешать при этом выборку (shuffle=True),
поскольку данные в таблице отсортированы по времени,
и без перемешивания можно столкнуться с нежелательными эффектами при оценивании качества.
Оцените качество градиентного бустинга (GradientBoostingClassifier)
с помощью данной кросс-валидации,
попробуйте при этом разное количество деревьев
(как минимум протестируйте следующие значения для количества деревьев: 10, 20, 30).
Долго ли настраивались классификаторы? Достигнут ли оптимум
на испытанных значениях параметра n_estimators,
или же качество, скорее всего, продолжит расти при дальнейшем его увеличении?

```python
# определили разбиение
k_fold = KFold(n_splits=5, shuffle=True, random_state=241)
result = {}
for estimators_count in [10, 20, 30, 60, 100]:
    model = GradientBoostingClassifier(n_estimators=estimators_count, random_state=241)
    model.fit(X, y)
    # запуск таймера
    start_time = datetime.datetime.now()
    scores = cross_val_score(model, X, y, cv=k_fold, scoring='roc_auc')
    print('For {} estimators cross_val_score takes time: {}'.format(estimators_count, datetime.datetime.now() - start_time)) 
    result[estimators_count] = scores.mean()
    
pprint(result)
pprint(max(result.values()))
```

Кросс-валидация с 30 деревьями: среднее значение = 0.689, время выполнения = 0:00:58. Использовать более 30 деревьев оправдано, но потребуется больше времени и значительного увеличения кол-ва деревьев - т.к даже при 100 получается достаточно невысокое значение эффективности классификатора - 0.706. Для ускорения обучения при увеличении количества деревьев можно использовать только часть выборки, уменьшение глубины дерева, более тонкая настройка параметров.

# Подход 2: логистическая регрессия

1. Оцените качество логистической регрессии (sklearn.linear_model.LogisticRegression с L2-регуляризацией) с помощью кросс-валидации по той же схеме, которая использовалась для градиентного бустинга. Подберите при этом лучший параметр регуляризации (C). Какое наилучшее качество у вас получилось? Как оно соотносится с качеством градиентного бустинга? Чем вы можете объяснить эту разницу? Быстрее ли работает логистическая регрессия по сравнению с градиентным бустингом?

```python
scaler = StandardScaler()
X_logic = scaler.fit_transform(X)
k_fold = KFold(n_splits=5, shuffle=True, random_state=241)
result = {}
for C in np.power(10.0, np.arange(-5, 6)):
    model = LogisticRegression(penalty='l2', C=C, random_state=241)
    model.fit(X_logic, y)
    # запуск таймера
    start_time = datetime.datetime.now()
    scores = cross_val_score(model, X_logic, y, cv=k_fold, scoring='roc_auc')
    print('For C = {} cross_val_score takes time: {}'.format(C, datetime.datetime.now() - start_time)) 
    result[C] = scores.mean()
pprint(result)
pprint(max(result.values()))
```

Уже при С=0.01 достигнут максимум: 0.716. Качество в целом совпадает с качеством градиентного бустинга - но в отличие от него обучение выполняется за значительно меньшее время.
2. Среди признаков в выборке есть категориальные, которые мы использовали как числовые, что вряд ли является хорошей идеей. Категориальных признаков в этой задаче одиннадцать: lobby_type и r1_hero, r2_hero, ..., r5_hero, d1_hero, d2_hero, ..., d5_hero. Уберите их из выборки, и проведите кросс-валидацию для логистической регрессии на новой выборке с подбором лучшего параметра регуляризации. Изменилось ли качество? Чем вы можете это объяснить?

```python
removed_columns_with_category = [
    'lobby_type', 
    *['r{}_hero'.format(i) for i in range(1,6)],
    *['d{}_hero'.format(i) for i in range(1,6)]
]
X_logic_without_category = X[[col for col in X.columns if col not in removed_columns_with_category]]
X_logic_without_category.head()
```

```python
scaler = StandardScaler()
X_logic_without_category = scaler.fit_transform(X_logic_without_category)
k_fold = KFold(n_splits=5, shuffle=True, random_state=241)
result = {}

for C in np.power(10.0, np.arange(-5, 6)):
    model = LogisticRegression(penalty='l2', C=C, random_state=241)
    model.fit(X_logic_without_category, y)
    # запуск таймера
    start_time = datetime.datetime.now()
    scores = cross_val_score(model, X_logic_without_category, y, cv=k_fold, scoring='roc_auc')
    print('For C = {} cross_val_score takes time: {}'.format(C, datetime.datetime.now() - start_time)) 
    result[C] = scores.mean()
pprint(result)
pprint(max(result.values()))
```

При С=0.01 достигнут максимум: 0.716. Качество предсказания осталось неизменным - т.к. в изначальном варианте алгоритм проигнорировал категориальные признаки.

3. На предыдущем шаге мы исключили из выборки признаки rM_hero и dM_hero, которые показывают, какие именно герои играли за каждую команду. Это важные признаки — герои имеют разные характеристики, и некоторые из них выигрывают чаще, чем другие. Выясните из данных, сколько различных идентификаторов героев существует в данной игре (вам может пригодиться фукнция unique или value_counts).

```python
with_category = X[[col for col in X.columns if col in removed_columns_with_category]]

heroes_unique = set()
for col in with_category.columns:
    heroes_unique.update(set(with_category[col].unique()))

heroes_count = max(heroes_unique)
print(heroes_count)
```

4. Воспользуемся подходом "мешок слов" для кодирования информации о героях. Пусть всего в игре имеет N различных героев. Сформируем N признаков, при этом i-й будет равен нулю, если i-й герой не участвовал в матче; единице, если i-й герой играл за команду Radiant; минус единице, если i-й герой играл за команду Dire. Ниже вы можете найти код, который выполняет данной преобразование. Добавьте полученные признаки к числовым, которые вы использовали во втором пункте данного этапа.

```python
# N — количество различных героев в выборке
X_pick = np.zeros((X.shape[0], heroes_count))
for i, match_id in enumerate(X.index):
     for p in range(5):
         X_pick[i, X.loc[match_id, 'r%d_hero' % (p+1)]-1] = 1
         X_pick[i, X.loc[match_id, 'd%d_hero' % (p+1)]-1] = -1
            
X_pick = pandas.DataFrame(X_pick, index=X.index)
```

```python
X_logic_without_category = X[[col for col in X.columns if col not in removed_columns_with_category]]
scaler = StandardScaler()
X_logic_without_category=pandas.DataFrame(scaler.fit_transform(X_logic_without_category), index=X_logic_without_category.index)
X_logic = pandas.concat([X_logic_without_category, X_pick], axis=1)
```

5. Проведите кросс-валидацию для логистической регрессии на новой выборке с подбором лучшего параметра регуляризации. Какое получилось качество? Улучшилось ли оно? Чем вы можете это объяснить?

```python
k_fold = KFold(n_splits=5, shuffle=True, random_state=241)
result = {}
for C in np.power(10.0, np.arange(-5, 6)):
    model = LogisticRegression(penalty='l2', C=C, random_state=241)
    model.fit(X_logic, y)
    # запуск таймера
    start_time = datetime.datetime.now()
    scores = cross_val_score(model, X_logic, y, cv=k_fold, scoring='roc_auc')
    print('For C = {} cross_val_score takes time: {}'.format(C, datetime.datetime.now() - start_time)) 
    result[C] = scores.mean()
pprint(result)
pprint(max(result.values()))
```

Качество улучшилось - при C = 1 получен максимум 0.751. Связано с более полными данными, 
которые дополнили сведениями о героях, что и улучшило работу предсказания.
